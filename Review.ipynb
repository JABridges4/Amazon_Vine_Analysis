```python
import pandas as pd
```

```python
vine_df= pd.read_csv('vine_table.csv')
vine_df.head(5)
```

```python
counts_vine_df= vine_df[vine_df["total_votes"]>=20]
counts_vine_df.head(5)
```

```python
helpful_df= counts_vine_df[counts_vine_df["helpful_votes"]/ counts_vine_df["total_votes"]>=.5]
helpful_df.head(5)
```

```python
helpful_vine_df= helpful_df[helpful_df["vine"]== 'Y']
helpful_vine_df.head(5)
```

```python
helpful_vine_unpaid_df= helpful_df[helpful_df["vine"]== 'N']
helpful_vine_unpaid_df.head(5)
```
